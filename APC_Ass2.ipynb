e) Looking at the plotted graphs for c and d, for the winter, we can see pick concentrations of CO between 9am and 10am, we can also notice peak concentrations between 11pm - 1am. During the summer, we can notice peak concentrations between 8am - 10am and also between 7pm to 9pm. It is observed that CO concentrations are higher in the winter than in the summer with peak periods skewed towards Saturdays (however, this sample does not account for Sunday - Tuesday). Time changes in peak period could be attributed to vehicular emissions and culture changes, as people tend to go to work earlier in the summer compared to winter. As cold air is denser than warm air, cold air traps CO emissions with less dispersion (remains in place for much longer), which could explain the higher concentration in the winter.
